#### Define

Our first tidy issue is that the tweet_data dataframe is not its own observational unit, but rather it contains variables that pertain to the tweets in the original tweets dataframe. We must, therefore, migrate the relevant columns from the tweet_data dataframe to the tweets dataframe. I will first add new columns for retweet_count and favorite_counts as those are the only variables we pulled from the json. I will then need to match up row information based on the tweet_id. This would be a lot easier if the dataframes were indexed by the tweet_id as opposed to the numerical indexing. Therefore, I will also tackle the issue of changing the index to be the tweet_id for all three dataframes. 

#### Code

```python
#tweet_data is not its own observational unit, it merely contains info on the tweets which is already an observational unit in the original tweets dataframe
#https://stackoverflow.com/a/16327135
#adding new columns for retweet_count and favorite count
tweets_clean['retweet_count'] = 0
tweets_clean['favorites_count'] = 0
#check that the columns have been added
tweets_clean.head()
```

```python
#tweets index is numerical instead of linked to tweet_id in all 3 dataframes
#change index to tweet_id
tweets_clean.set_index(tweets_clean.columns[0], inplace = True)
tweets_clean.head()
```

```python
#tweets index is numerical instead of linked to tweet_id in all 3 dataframes
#change index to tweet_id
images_clean.set_index(images_clean.columns[0], inplace = True)
images_clean.head()
```

```python
#tweets index is numerical instead of linked to tweet_id in all 3 dataframes
#change index to tweet_id
tweet_data_clean.set_index(tweet_data_clean.columns[0], drop = False, inplace = True)
tweet_data_clean.head()
```

```python
#for all tweets, add tweet_data for retweet_count and favorite count to appropriate column

for x in tweets_clean.index.values:
    #if the tweet is deleted and therefore has no data this will throw an error
    try:
        #pull retweet count from tweet_data dataframe
        rc = int(tweet_data_clean.at[x, 'retweet_count'])
        #pull favorite count from tweet_data dataframe
        fc = int(tweet_data_clean.at[x, 'favorite_count'])
        #print statement for clarity
        print('retweets: ' + str(rc) + ' favorites: ' + str(fc))
        #add data to tweets dataframe
        tweets_clean.at[x, 'retweet_count'] = rc
        tweets_clean.at[x, 'favorites_count'] = fc
    except:
        #print statement to give a rough estimate of # of deleted tweets
        print('deleted tweet')
```

#### Define

Our second tidy issue is that the doggo, floofer, pupper, and puppo columns are all values of the dog classification variable. In order to resolve this I will make a new column called dog_class and use the doggo, floofer, pupper, and puppo columns to populate it. I'm sure that there is an elegant way to do this with merge or melt functions but I have a working solution using chained if else statements. I'm also not sure if there's any rows that have more than one dog class. If that is the case I will choose an order of precedence of dog classification.

#### Code

```python
#checking column names
tweets_clean.columns
```

```python
#make new column
tweets_clean['dog_class'] = None
```

```python
#check doggo count
tweets_clean.doggo.value_counts()
```

```python
#check floofer count
tweets_clean.floofer.value_counts()
```

```python
#check pupper count
tweets_clean.pupper.value_counts()
```

```python
#check puppo count
tweets_clean.puppo.value_counts()
```

```python
#in tweets table, [doggo, floofer, pupper, puppo] are all values instead of variables

#for each row
for x in tweets_clean.index.values:
    #if doggo is populated then update dog_class
    if tweets_clean.at[x,'doggo'] != 'None':
        tweets_clean.at[x,'dog_class'] = 'doggo'
    #if floofer is populated then update dog_class
    if tweets_clean.at[x,'floofer'] != 'None':
        tweets_clean.at[x,'dog_class'] = 'floofer'
    #if pupper is populated then update dog_class
    if tweets_clean.at[x,'pupper'] != 'None':
        tweets_clean.at[x,'dog_class'] = 'pupper'
    #if puppo is populated then update dog_class
    if tweets_clean.at[x,'puppo'] != 'None':
        tweets_clean.at[x,'dog_class'] = 'puppo'
```

The solution in the previous cell gives dog class precedence from bottom to top. So puppo > pupper > floofer > doggo.

#### Test

```python
tweets_clean.dog_class.value_counts()
```

Here we can see there is a mismatch of the doggo count in dog_class and those present in the original dataframe. This likely means that there are rows where both doggo is populated and another dog_class is populated. I will assume that doggo is the most general class of dog and therefore in the interest of specificity give precedence to all other classes over doggo.

```python
tweets_clean[tweets_clean.doggo == 'doggo']
```

We can see that this is indeed the case, there are rows where doggo is populated and another class is also populated.

Now we can drop the individual dog class columns as they are extraneous.

```python
tweets_clean.drop(columns = ['doggo','floofer','pupper','puppo'], inplace = True)
tweets_clean.head()
```

```python
#checking that columns have been dropped
tweets_clean.info()
```

#### Define

Now that we have all the dog classes grouped under the same column. We can now change the datatype to be the category using the astype function.

#### Code

```python
#the dog classification should be category data type instead of a string 
#https://pandas.pydata.org/pandas-docs/stable/categorical.html

tweets_clean['dog_class'] = tweets_clean['dog_class'].astype('category')
```

Similar to the above issue, timestamp is a string in our dataframe and making it into a datetime object would allow use to use all the functions associated with that object. There is a to_datetime function that is included with the pandas package that we can take advantage of.
